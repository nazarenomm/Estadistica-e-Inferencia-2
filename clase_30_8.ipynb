#### Cuan informativos deben ser los priors?

Es aconsejable usar priors no informativos, pero hay que tener cuidado con las observaciones imposibles que se pueden generar. Siempre que podemos obtener rangos mejor, por ejemplo, sabemos que un la masa de un organismo siempre va a estar entre la celulca mas pequea y la ballena azul. Lo mismo con promedios y dems.   

Los priors informativos estan buenos cuando tenes informacion, muchas veces no hacen ka diferencia y eso puede ser una perdida de tiempo.  

#### Prios levemente informativos
Lo mejor siempre esta entre las dos anteriores.  
Igualmente no hay una manera concreta de saber cual prior es levemente informativo.  



+ gris: distribucion real de los datos.  
+ azul: podriamos decir que es un prior vago, ya que mucha de la densidad esta por fuera de los datos reales.  
+ amarillo: la densidad de la informacion esta concentrada practicamente en un punto, en este caso da la casualidad que coincide con los datos reales, podria estar en alguno de los extremos del grafico, lo que generara un conflicto. Si vamos a usar un prior informativo mejor que estemmos seguros de que sea correcto.  
+ celeste: levemente informativo, gran parte del prior coincide con los datos, reduce la chance de regiones ridiculas.  

en la practica no sabemos lo gris, entonces podemos usar valores de referencia y prior predictives.  



Se recomienda usar priors levemente informativos.
