### Regresion polinomial
La $X$ va siendo transformada a $X^2$, $X^3$, $X^m$, donde $m$ es el grado del polinomio.

Usar la notación poly

### Splines

Es la función más comunmente asociada a los GAMs.

Son como polinomios pero estan restringidos a ciertas areas.

Particiono mi espacio de datos y en cada espacio aplico un polinomio.

Una restriccion que tienen es que los polinomios deben tocarse (salvo en el caso constante, grado 0)
